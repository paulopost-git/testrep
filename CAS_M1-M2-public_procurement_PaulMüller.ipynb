```python
# getting the dataframe public purchase (pp) ready
import pandas as pd
import numpy as np
#df_pp = pd.read_excel('Belegdatum_Rechnungsvolumen.xlsx',sheet_name='Bericht',header=4, index_col=0)
df_pp = pd.read_excel('DateInvoice.xlsx',sheet_name='Bericht',header=4, index_col=0)
df_pp.info()
df_pp.tail()
```

```python
from matplotlib import pyplot
df_pp.plot()
#pyplot.ylim(0, 100);
pyplot.title('Invoice volume for procurements')
pyplot.ylabel('Invoice volume in Mio CHF')
pyplot.show()
```

```python
df_pp_2 = df_pp.groupby(pd.Grouper(freq='M')).sum()
```

```python
df_pp_2.plot()
pyplot.title('Invoice volume for procurements')
pyplot.ylabel('Invoice volume in Mio CHF')
pyplot.show()
```

```python
# create a boxplot of monthly data
from pandas import Grouper
from pandas import concat
from pandas import DataFrame
one_year = df_pp['2019']
groups = one_year.groupby(Grouper(freq='M'))
months = concat([DataFrame(x[1].values) for x in groups], axis=1)
months = DataFrame(months)
months.columns = range(1,13)
months.boxplot()
pyplot.title('Avarage monthly invoice volume for procurements')
pyplot.ylabel('Invoice volume in Mio CHF')
pyplot.xlabel('Months')
pyplot.show()
```

```python
#df_pp_f = pd.read_excel('Belegdatum_Rechnungsvolumen_nach_Beschaffungsverfahren.xlsx',sheet_name='Bericht',header=4, index_col=0)
df_pp_f = pd.read_excel('DateInvoice_Procedure.xlsx',sheet_name='Bericht',header=4, index_col=0)
df_pp_f.info()
df_pp_f.head()
```

```python
df_pp_f = df_pp_f.fillna('0')
df_pp_f = df_pp_f.drop(['Buchungsdatum im Beleg'])
df_pp_f.head()
```

```python
df_pp_f = df_pp_f.apply(pd.to_numeric, errors='coerce').fillna(0, downcast='infer')
df_pp_f.dtypes
```

```python
df_pp_f['limited procedure'] = df_pp_f['Verg.unter WTO-Schwellw: freih.Verf.'] + df_pp_f['Verg.über WTO-Schwellw: freih.Verf.']
df_pp_f['competition procedures'] = df_pp_f['Verg.unter WTO-Schwellw: offen.Verf.'] + df_pp_f['Verg.über WTO-Schwellw: offen.Verf.'] + df_pp_f['Verg.unter WTO-Schwellw: selekt.Verf.'] +  df_pp_f['Verg.über WTO-Schwellw: selekt.Verf.'] +  df_pp_f['Verg.unter WTO-Schwellw: Einlad.Verf.'] + df_pp_f['Verg.über WTO-Schwellw: Einlad.Verf.']
df_pp_f = df_pp_f.drop(df_pp_f.columns[[0, 1, 2, 3, 4, 5, 6, 7]], axis=1)
df_pp_f.head()
```

```python
df_pp_f2 = df_pp_f.groupby(pd.Grouper(freq='M')).sum()
```

```python
df_pp_f2.plot()
pyplot.title('Invoice volume for procurements')
pyplot.ylabel('Invoice volume in Mio CHF')
pyplot.xlabel('Months')
pyplot.show()
```

```python
df_pp_f['Rvol'] = df_pp_f['limited procedure'] + df_pp_f['competition procedures']
df_pp_f['Anteil_f'] = df_pp_f['limited procedure'] / df_pp_f['Rvol']
df_pp_f.head()
```

```python
df_pp_f.mean()
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(df_pp_f.Anteil_f, df_pp_f.Rvol)
slope, intercept, r_value, p_value, std_err
```

```python
df_pp_f = df_pp_f.drop(df_pp_f.columns[[0, 1, 2]], axis=1)
df_pp_f.plot()
pyplot.show()
```

```python
stats.ks_2samp(df_cp123['1st to 3rd quarter'][:40],df_cp4['4th quarter'][:40]) #the Kolmogorov-Smirnov statistic on 2 samples. This is a test for the null hypothesis that 2 independent samples are drawn from the same continuous distribution.
```
